```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
from torch.distributions import Normal, Uniform, Gamma
```

```python
target_distribution = Gamma(concentration=5, rate=1)
# target_sample = target_distribution.sample([128])
x = np.linspace(0, 40, 1000)
plt.plot(x, target_distribution.log_prob(torch.tensor(x)).exp().tolist(), label='Target')
plt.legend()
# plt.xticks(np.linspace(0.001, 20, 1))
```

```python
def nlp_pdf(x, tau):
    return x**2*1/tau*(1/np.sqrt(2*np.pi*tau))*np.exp(-x**2/(2*tau))
x = np.linspace(0, 10, 1000)
x_ext = np.linspace(-10, 10, 1000)
plt.plot(x_ext,nlp_pdf(x_ext, tau=6), label='NLP')

gamma = Gamma(concentration=4, rate=1.05)
plt.plot(x, (0.5*gamma.log_prob(torch.tensor(x)).exp()).tolist(), label='Gamma')

normal = Normal(loc=3, scale=1.7)
plt.plot(x_ext, (0.5*normal.log_prob(torch.tensor(x_ext)).exp()).tolist(), label='Normal')

plt.axvline(x=0, linestyle='--')

plt.legend()
    
```
